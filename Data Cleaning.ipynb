CLEANING OF DATASET

FINDING NULL VALUES IN DATASET:

```python
df.isnull().sum()#to check if there are any null values in the dataset
```

```python
df = df.dropna(subset="mileage")#to drop the null values in the mileage column
df = df.dropna(subset="torque")#to drop the null values in the torque column
df
```

FINDING DUPLICATE VALUES IN THE DATASET:

```python
dup = df.duplicated().sum()#to check if there are any duplicate values in the dataset
dup
```

```python
df =df.drop_duplicates()#to drop the duplicate values in the dataset
df
```

RE-CHECKING NULL VALUES IN DATASET:

```python
df.isnull().sum()#to check if there are any null values in the dataset
```

```python

df[['engine', 'engine_unit']] = df['engine'].str.split(' ', expand=True)#to split the engine column into two columns, one for the engine value and one for the engine unit
```

```python
df
```

```python
df[['max_power', 'max_power_unit']] = df['max_power'].str.split(' ', expand=True)#to split the max_power column into two columns, one for the max_power value and one for the max_power unit
df
```

```python
df[['mileage', 'mileage_unit']] = df['mileage'].str.split(' ', expand=True)#to split the mileage column into two columns, one for the mileage value and one for the mileage unit
df
```
