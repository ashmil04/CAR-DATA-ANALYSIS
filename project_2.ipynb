# Interesting question/problem statements and thier findings

1.how does the car age affect selling price?


- answerd this question by showing the scatter plot and line graph(using the groupby funtion)of the category of price and year of the car

```python
year_price=df.groupby("year")["selling_price"].mean()
plt.plot(year_price.index,year_price.values,marker= 'o')
plt.title('year vs selling price')
plt.xlabel("year")
plt.ylabel("average price")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["year"], df["selling_price"], alpha=0.5)
plt.xlabel("Manufacturing Year")
plt.ylabel("Selling Price")
plt.title("Relationship between Car Age and Selling Price")
plt.grid(True)
plt.show()
```

- Negative relationship
→ As car age increases, selling price decreases.
- Newer cars have higher prices
→ Recently manufactured cars are more expensive.
- Older cars have lower prices
→ Older vehicles are sold at cheaper rates.
- Depreciation effect
→ Car value reduces over time due to usage and wear.
- Higher price variation in newer cars
→ Prices of newer cars vary widely based on features and condition.
- Lower price concentration in older cars
→ Older cars mostly fall within a lower price range

# 3.Which fuel type cars have the best resale value?

```python
fuel_price = df.groupby("fuel")["selling_price"].mean().reset_index()
sns.barplot(x= "fuel",y= "selling_price",data= fuel_price)
plt.xlabel("fuel type")
plt.ylabel("average selling price")
plt.title("fuel type vs selling price ")
plt.show()
```

- Diesel cars have the best resale value,followed by petrol,while CNG and LPG have lower resale value
- Diesel > Petrol > LPG And CNG

# 4.Do automatic cars sell for higher prices than manual cars?

```python
trans_price = df.groupby("transmission")["selling_price"].mean().reset_index()
sns.barplot(x ="transmission", y = "selling_price",data =  trans_price)
plt.xlabel("transmission")
plt.ylabel("average selling price")
plt.title("transmission vs selling price")
```

- Yes , Automatic transsion has more resalue value than Manual transmission


# 5.Can we predict selling price using features like year, km_driven, engine, and fuel?

yes, we can predict the selling price by using features like year , km_driven engine and fuel

```python
corr = df[["year", "km_driven", "engine", "selling_price"]].corr()
print(corr)
```

```python
sns.heatmap(corr, annot=True)
plt.title("Correlation Matrix")
plt.show()
```

- As the year goes new the the selling price goes up that is new car -> high price
- The no kilometers driven increses the price decreases that is higher the km -> the price goes down
- As the engine capacity high the the price goes up the price goes up
- Fuel type has a moderate influence on selling price, with diesel cars generally having higher resale value
